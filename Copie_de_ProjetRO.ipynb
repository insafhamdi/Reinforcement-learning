# **<font color='tiffany bleu'>R√©solution du PL : </font>**<br>


```python
import numpy as np
from scipy.optimize import linprog

# Fixer la graine al√©atoire pour la reproductibilit√©
np.random.seed(123)

# D√©claration des dimensions
Nue = 10
Ndu = 5
Nvsf = 3
Nnet = 7
Nsins = 2
Nucls = [1, 2, 3]
Ereq = [(1, 2), (2, 3)]

# G√©n√©ration de valeurs al√©atoires pour c
c = np.random.rand(Nue * Ndu + Nue * Nvsf * Nnet * Nsins + Nvsf * Nnet * Nsins + Nue * Nue * Nnet + Nue * len(Ereq))

# G√©n√©ration de valeurs al√©atoires pour A_ub et b_ub
A_ub = np.random.rand(5, len(c))
b_ub = np.random.rand(5)

# G√©n√©ration de valeurs al√©atoires pour A_eq et b_eq
A_eq = np.random.rand(Nue, len(c))
b_eq = np.random.rand(Nue)

# D√©finition des bornes des variables
x_bounds = [(0, 1) for _ in range(len(c))]

# R√©solution du probl√®me lin√©aire avec la m√©thode du simplexe
result = linprog(c, A_ub=A_ub, b_ub=b_ub, A_eq=A_eq, b_eq=b_eq, bounds=x_bounds, method='simplex')



# Affichage des r√©sultats
print("Optimal value:", result.fun)
print("Optimal xi values:")
print(result.x)
```
